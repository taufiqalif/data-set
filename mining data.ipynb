```python
# Instal dulu snscrape (hanya sekali)
!pip install snscrape

# Import library
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Keyword pencarian
query = "hoaks OR berita palsu OR informasi salah lang:id"
max_tweets = 100

# List penampung hasil scraping
tweets = []

# Proses scraping
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    if len(tweets) == max_tweets:
        break
    tweets.append([tweet.date, tweet.user.username, tweet.content])

# Simpan ke DataFrame
df_scraped = pd.DataFrame(tweets, columns=["tanggal", "username", "tweet"])

# Simpan ke CSV
df_scraped.to_csv("hasil_scraping_twitter.csv", index=False)

print(df_scraped.head())
```
